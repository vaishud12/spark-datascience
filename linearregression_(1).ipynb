**step1:import the dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading data from remote link using the url
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
student_data = pd.read_csv(url)

print("Data imported successfully")
student_data
```

```python
student_data.shape
#here we can see that there are 25 rows and 2 coloums in the dataset
```

```python
student_data.describe()
```

```python
student_data.isnull().sum()
#here we can see that there are no null values in the dataset that can affect the training of our algorithm
```

### **step2: Visualize and Analyze the Dataset**


```python
#plotting the distribution of scores and number of hours studied on a 2D graph
student_data.plot(x='Hours', y='Scores', style='ro')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage score')
plt.show()
```

from the above graph we can see that there is a positive linear relationship between Hours and percentage with means that as the number of Noun studied
 

**step3:Prepare the Data**

```python
# we are extracting values of Hours Data into variable x and the values of scores Data into variable Y
x = student_data.iloc[:, :-1].values
y = student_data.iloc[:, 1].values
```

```python
#number of hours studied
x
```

```python
#scores obtained
y
```

```python
# we now split the data into train and test the dataset using scikit-learn's built-in train_test_split()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                               test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**step4:Design and Train the machine learning model**

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)
print("Training complete.")
```

**step5:visualize the model**

```python
#plotting the regression line
# Plotting the regression line
line = regressor.coef_*x+regressor.intercept_

# Plotting for the test data
plt.scatter(x, y)
plt.plot(x, line, color="yellow");
plt.show()
```

**step6:Make predictions**

```python
print(x_test) # testing  data - In Hours
y_pred = regressor.predict(x_test) # predicting the scores
```

```python
#comaparing actual vs predicted
df = pd.DataFrame({'Actual': y_test, 'predicted': y_pred})
df
```

```python
# You can also test with your own data
hours = 9.25
own_pred = regressor.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

**step7:Evaluate the model**

**mean error**

```python
from sklearn import metrics
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square error. There are many such metrics.

**max error**

```python
from sklearn import metrics
print('Max Absolute Error:', 
      metrics.max_error(y_test, y_pred)) 
```

**mean squared error**

```python
from sklearn import metrics
print('Mean squared Error:', 
      metrics.mean_squared_error(y_test, y_pred)) 
```
